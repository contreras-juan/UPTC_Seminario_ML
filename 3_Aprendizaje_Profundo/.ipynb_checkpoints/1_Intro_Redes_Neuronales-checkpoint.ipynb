<h1 align='center'> Introducción a las redes neuronales artificiales </h1>

Las redes neuronales artificiales son sistemas computacionales inspirados en el funcionamiento del cerebro humano. Se componen de unidades interconectadas llamadas neuronas, organizadas en capas, que trabajan juntas para procesar información. Estas redes son capaces de aprender y mejorar a partir de datos, lo que las hace especialmente útiles para tareas complejas como el reconocimiento de patrones, la clasificación y la predicción. A través de técnicas de aprendizaje profundo, las redes neuronales pueden manejar grandes volúmenes de datos y extraer características relevantes de manera automática, revolucionando campos como la inteligencia artificial, la visión por computadora y el procesamiento del lenguaje natural.

<h2> Redes neuronales biológicas </h2>

<img src='img/neurona.jpg'>

Fuente: [Universidad Estatal a Distancia](https://multimedia.uned.ac.cr/pem/anatomia_especies_silvestres/pant/nervioso/neurona.html)

**Axón:** Su función principal es conducir los impulsos nerviosos (señales eléctricas) desde el cuerpo celular hacia otras neuronas, músculos o glándulas. Los axones están frecuentemente cubiertos por una capa de mielina, una sustancia aislante que aumenta la velocidad de transmisión de los impulsos nerviosos. La mielina está formada por células de Schwann en el sistema nervioso periférico y por oligodendrocitos en el sistema nervioso central. El extremo del axón se ramifica en múltiples terminales axónicos, que hacen contacto con otras células en las sinapsis.

**Núcleo Celular:** El núcleo celular se encuentra dentro del cuerpo celular de la neurona y contiene el material genético (ADN) de la célula. El núcleo dirige todas las actividades celulares, incluyendo el crecimiento, la reproducción y la síntesis de proteínas necesarias para el funcionamiento y mantenimiento de la neurona.

**Dendritas:** Las dendritas son prolongaciones ramificadas que se extienden desde el cuerpo celular de la neurona. Su principal función es recibir señales de otras neuronas y transmitirlas hacia el cuerpo celular.

**Cuerpo Celular:** El cuerpo celular es responsable de la integración de las señales recibidas por las dendritas y de la generación de nuevos impulsos nerviosos que serán transmitidos a lo largo del axón. Además, el soma también participa en el mantenimiento y reparación de la neurona.

### Funcionamiento General de la Neurona
El proceso de transmisión de la información en una neurona se puede describir de la siguiente manera:

1. **Recepción de señales:** Las dendritas reciben señales químicas de otras neuronas a través de las sinapsis. Estas señales químicas se transforman en señales eléctricas en la membrana de la dendrita.
2. **Integración de señales:** Las señales eléctricas recibidas por las dendritas se suman en el cuerpo celular. Si la señal total alcanza un cierto umbral, se genera un potencial de acción en el segmento inicial del axón.
3. **Conducción del impulso nervioso:** El potencial de acción se propaga a lo largo del axón hacia sus terminales. La conducción es facilitada por la mielina en los axones mielinizados, que permite saltos rápidos de los impulsos entre los nodos de Ranvier (interrupciones en la mielina).
4. **Transmisión de señales:** Cuando el impulso nervioso alcanza las terminales del axón, provoca la liberación de neurotransmisores en las sinapsis. Estos neurotransmisores cruzan la sinapsis y se unen a los receptores en las dendritas de la siguiente neurona, reiniciando el ciclo.

<h2> Redes neuronales artificiales </h2>

<img src='img/red_neuronal_art.png'>

Fuente: [Meng et al (2020)](https://www.researchgate.net/figure/A-biological-neuron-in-comparison-to-an-artificial-neural-network-a-human-neuron-b_fig2_339446790)

Anatomía de una red neuronal artifical:

1. **Capa de entrada**: Es la primera capa de una red neuronal artificial donde se introducen los datos de entrada. Cada nodo en esta capa representa una característica o variable de entrada y transmite esta información a los nodos de la capa siguiente.

2. **Neurona**: En el contexto de una red neuronal artificial, una neurona es una unidad computacional básica que procesa información. Toma entradas ponderadas, las suma, aplica una función de activación y produce una salida.

3. **Capa oculta**: Una capa oculta es cualquier capa de una red neuronal artificial que no es ni la capa de entrada ni la capa de salida. Estas capas realizan cálculos intermedios para aprender y extraer características complejas de los datos de entrada.

4. **Capa de salida**: Es la última capa de una red neuronal artificial que produce los resultados finales o las predicciones basadas en las entradas y los cálculos realizados en las capas anteriores. Cada nodo en esta capa representa una posible salida o clase.

5. **Función de activación:** Es una función matemática aplicada a la salida de cada neurona en una red neuronal artificial. Esta función introduce no linealidades en la red y ayuda a la red a aprender y modelar relaciones complejas en los datos de entrada.

6. **Predicción**: Es el resultado generado por una red neuronal artificial para un conjunto dado de entradas. En el contexto del aprendizaje supervisado, la predicción es la salida estimada de la red neuronal para una entrada dada después de haber sido entrenada con ejemplos etiquetados.

Sea $\mathbf{X}_{N \text{x} K}$ una matriz de datos:

$$\mathbf{X} = [\mathbf{x}_{1}, \mathbf{x}_{2}, ..., \mathbf{x}_{k}] =  \begin{pmatrix}
x_{11} & x_{12} & \cdots &  x_{1k} \\
x_{21} & x_{22} & \cdots & x_{2k} \\
\vdots & \cdots & \ddots & \vdots \\
x_{N1} & x_{N2} & \cdots & x_{nK} \\
\end{pmatrix} $$

y $y_{N \text{x} 1}$ un vector de salida:

$$y = \left(y_{1}, y_{2}, ..., y_N \right)^{T}$$

construimos un modelo matemático tal que $\mathbf{\hat{y}} = f\left(\mathbf{X} \right)$ en la que cada columna $x_{j}$ es multiplicada por un peso $w_{j}$.

Comencemos por una red neuronal sencilla en la que queremos desarrollar un ejercicio de regresión con base en $K$ variables para $N$ observaciones. Esto significa:

- $\mathbf{X}:$ matriz de entrada de dimensiones $N \text{x} K$

- $\mathbf{W}_{1}:$ matriz de dimensiones $K \text{x} 2$ de pesos que conecta la capa de entrada con la primera capa oculta.

- $\mathbf{Z}_{1}:$ vector $N \text{x} 2$ que corresponde a los valores en las neuronaes de la capa oculta

- $A_{1}$: valor de activación, que corresponde al valor de las neuronas de la primera capa oculta al aplicar la función de activación, esto es $A_{1} = f(Z_{1})$

- $W_2$: matriz $2 \text{x} 1$ de pesos que conecta la capa oculta con la capa de salida

- $Z_{2}:$ valor que corresponde al producto $Z_{2} = Z_{1} W_{2}$

- $\hat{y}:$ predicción de la variable $y$, $\hat{y} = f_{2} (Z_{2})$

Una red como estas tiene forma como la de la siguiente imagen:

<img src='img/ex_ANN.png' width=1000>

Fuente: [Yamamoto et al (2018)](https://www.researchgate.net/figure/Artificial-Neural-Network-with-two-inputs-and-a-single-hidden-layer-Each-artificial_fig1_324615440)

De forma resumida tenemos:

$$Z_{1} = \mathbf{X} \mathbf{W}_{1} + b_{1}$$
$$A_{1} = f_{1}(Z_{1})$$
$$Z_{2} = \mathbf{A}_{1} \mathbf{W}_{2} + b_{2}$$
$$\hat{y} = f_{2}(Z_{2})$$

<h2> Funciones de activación </h2>

Si las funciones de activación que escogemos son lineales, el modelo que tenemos es esencialmente una regresión lineal. Sin embargo, lo interesante es escoger funciones no lineales con el objetivo de aprender patrones más complejos de los datos.

<h3> ReLu </h3>

La función de activación ReLU (Rectified Linear Unit) es una de las funciones de activación más populares y ampliamente utilizadas en redes neuronales, especialmente en redes neuronales profundas (deep learning). Su popularidad se debe a varias ventajas que ofrece sobre otras funciones de activación.

$$ReLu = max\left(0, x \right)$$

**Algunas ventajas de ReLU**

- No linealidad: Aunque parece una función lineal, ReLU introduce no linealidad en el modelo, permitiendo que la red neuronal aprenda representaciones complejas.

- Computacionalmente eficiente: ReLU es simple de calcular, ya que solo implica una comparación y no requiere operaciones exponenciales, lo que la hace computacionalmente eficiente.

- Sparsity (Esparsidad): Debido a que la salida de ReLU es 0 para cualquier valor negativo, esto crea una red neuronal que es esparsa, es decir, muchas neuronas no se activan simultáneamente. Esto puede llevar a modelos más eficientes y a una mejor generalización.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)
```

```python
# Generar datos de entrada
x = np.linspace(-10, 10, 100)
y = relu(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ReLU(x)', color='blue')
plt.title('Función de Activación ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()
plt.grid(True)
plt.show()
```

<h3> $\tanh$ </h3>

La función de activación $\tanh$, que es la abreviatura de tangente hiperbólica, es una función matemática ampliamente utilizada en redes neuronales. Su fórmula es:

$$\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$

Esta función transforma una entrada escalar \( x \) en un valor en el rango $[-1, 1]$.

### Características de $\tanh$:

1. **Rango**: La salida de $\tanh(x)$ varía entre $-1$ y $1$.
2. **Simetría**: La función $\tanh$ es simétrica respecto al origen (función impar), es decir, $\tanh(-x) = -\tanh(x)$.
3. **Punto de equilibrio**: La función pasa por el origen, es decir, $\tanh(0) = 0$.
4. **Saturación**: Para valores grandes positivos de $x$, $\tanh(x)$ tiende a 1, y para valores grandes negativos de $x$, $\tanh(x)$ tiende a -1. Esto significa que la función se "satura" o aplana para valores extremos de $x$.

### Ventajas de usar $\tanh$:

1. **Salida centrada en cero**: Las salidas de las neuronas pueden ser negativas, lo cual ayuda a centrar los datos y facilita el entrenamiento de la red. Las salidas centradas en cero pueden llevar a convergencias más rápidas ya que la función de error puede oscilar alrededor de cero.

2. **Gradientes más grandes**: La \( \tanh \) tiene gradientes más grandes que la sigmoide en su rango principal, lo que puede ayudar a reducir problemas de gradiente desaparecido durante el entrenamiento de redes profundas.

### Desventajas de usar $\tanh$:

1. **Saturación**: Similar a la función sigmoide, $\tanh$ puede sufrir de saturación. Cuando los valores de entrada son muy grandes o muy pequeños, los gradientes se vuelven muy pequeños (se acercan a cero), lo que puede llevar a un aprendizaje muy lento (problema del gradiente desaparecido).

```python
def tanh(x):
    return np.tanh(x)
```

```python
# Generar datos de entrada
x = np.linspace(-10, 10, 100)
y = tanh(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='tanh(x)', color='blue')
plt.title('Función de Activación tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.legend()
plt.grid(True)
plt.show()
```

<h3> Sigmoide </h3>

La función de activación sigmoide es una de las funciones de activación más utilizadas en redes neuronales, especialmente en las capas ocultas y de salida de los modelos clásicos de aprendizaje automático. La función de activación sigmoide tiene la forma matemática:

$\sigma(x) = \frac{1}{1 + e^{-x}}$

### Características de la función sigmoide

1. **Salida en el rango (0, 1)**: La función sigmoide convierte cualquier valor real de entrada en un valor en el rango de 0 a 1. Esto es útil para modelar probabilidades, ya que una salida de 0.5 puede interpretarse como una probabilidad del 50%.

2. **No linealidad**: La sigmoide introduce no linealidad en el modelo, permitiendo a la red neuronal aprender representaciones complejas de los datos. Sin funciones no lineales, una red neuronal profunda simplemente sería equivalente a una sola capa lineal.

3. **Diferenciabilidad**: La función sigmoide es continuamente diferenciable, lo que es una propiedad crucial para los algoritmos de aprendizaje basados en gradientes, como el descenso de gradiente. La derivada de la función sigmoide se puede calcular de manera sencilla y eficiente.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Generar valores de x en el rango de -10 a 10
x_values = np.linspace(-10, 10, 100)

# Calcular los valores de y utilizando la función sigmoide
y_values = sigmoid(x_values)

# Graficar la función sigmoide
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Función Sigmoide', color='blue')
plt.title('Gráfico de la Función Sigmoide')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

Esta función de activación se utiliza en las capas de salida de los problemas de clasificación binaria.

<h3> Softmax </h3>

La función de activación softmax es comúnmente utilizada en redes neuronales, especialmente en la capa de salida de clasificación multiclase. Su función principal es transformar un vector de números reales en un vector de probabilidades. Esto es útil en tareas de clasificación donde se necesita asignar una probabilidad a cada clase posible.

La función softmax opera tomando como entrada un vector de números reales $z$ y produce como salida un vector de la misma dimensión donde cada elemento está en el rango $[0, 1]$ y la suma de todos los elementos es igual a 1. Esto se logra mediante la siguiente fórmula:

$$\text{softmax}(z)_i = \frac{e^{z_i}}{\sum_{j=1}^{n} e^{z_j}}$$

Donde:
- $z_i$ es el $i$-ésimo elemento del vector de entrada.
- $\text{softmax}(z)_i$ es el $i$-ésimo elemento del vector de salida después de aplicar softmax.
- $e$ es la constante de Euler (aproximadamente igual a 2.71828).
- $n$ es la dimensión del vector $z$.

La función softmax tiene varias propiedades útiles:

1. **Escalado de valores**: Los valores de salida están escalados en el rango $[0, 1]$, lo que los hace interpretables como probabilidades.

2. **Difuminación de la atención**: La softmax aumenta la diferencia entre los valores de salida, lo que significa que una entrada con un valor relativamente mayor se convierte en una probabilidad mucho mayor, mientras que una entrada con un valor relativamente menor se convierte en una probabilidad mucho menor.

3. **Diferenciabilidad**: Esta función es diferenciable, lo que es crucial para el entrenamiento de redes neuronales mediante técnicas como el descenso de gradiente.

Esta función de activación se utiliza en las capas de salida de los problemas de clasificación multiclase.

<h2> Teorema de Aproximación Universal </h2>

> 
> Sea $\sigma (\bullet)$ una función de activación continua y no constante. Entonces, para cualquier función continua $f(\bullet)$ definida en un intervalo compacto $[a, b] \subset \mathbb{R}^{n}$ y para cualquier $\epsilon > 0$, existe una red neuronal de una sola capa oculta con un número finito de neuronas, cuyos pesos y sesgos pueden ser ajustados de tal manera que la red aproxime $f(\bullet)$ con un error menor que $\epsilon$.

El Teorema de Aproximación Universal fue demostrado por primera vez por [George Cybenko en 1989](https://link.springer.com/article/10.1007/BF02551274) para funciones de activación sigmoide y más tarde fue generalizado por otros investigadores para diferentes tipos de funciones de activación.

### Implicaciones del Teorema

1. **Capacidad de Aproximación:**
   El teorema garantiza que incluso una red neuronal con una sola capa oculta tiene la capacidad teórica de aproximar cualquier función continua en un dominio finito. Esto resalta el poder y la flexibilidad de las redes neuronales como aproximadores universales de funciones.

2. **Necesidad de Profundidad vs. Ancho:**
   Si bien el teorema se refiere a redes neuronales de una sola capa oculta, en la práctica, las redes profundas (aquellas con múltiples capas ocultas) suelen ser más eficientes y requerir menos neuronas en total para lograr la misma precisión de aproximación. Esto se debe a que las redes profundas pueden capturar características más complejas de los datos de manera jerárquica.

3. **Función de Activación:**
   La elección de la función de activación es crucial. El teorema original de Cybenko se refería a funciones de activación sigmoides, pero se ha demostrado que otras funciones de activación como ReLU también cumplen con el teorema de aproximación universal.

### Limitaciones del Teorema

1. **Número de Neuronas:**
   Aunque el teorema garantiza la existencia de una red neuronal que puede aproximar cualquier función continua, no proporciona un límite superior práctico para el número de neuronas necesarias. En muchos casos, el número requerido puede ser imprácticamente grande.

2. **Funciones Continuas:**
   El teorema aplica únicamente a funciones continuas. Para funciones que no son continuas o que tienen singularidades, se requieren enfoques adicionales o diferentes tipos de redes.

3. **Entrenamiento:**
   El teorema es una afirmación teórica sobre la capacidad de representación de las redes neuronales. En la práctica, entrenar una red neuronal para que logre una aproximación adecuada puede ser un desafío debido a problemas como el sobreajuste, la optimización y la selección de hiperparámetros.
